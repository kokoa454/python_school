```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("data.csv", index_col = 0)
print(data)
```

```python
from sklearn.decomposition import PCA

model_pca = PCA(n_components = 2)
vecs_list = model_pca.fit_transform(data)
```

```python
X = vecs_list[:, 0]
y = vecs_list[:, 1]

plt.figure(figsize = (8, 8))

sns.scatterplot(x = X, y = y)

for i, (x_name, y_name) in enumerate(zip(X, y)):
    plt.annotate(data.index[i], (x_name, y_name))

plt.show()
```

```python
X_comp, y_comp = model_pca.components_

plt.figure(figsize = (8, 8))

sns.set(font = "Yu Gothic")
sns.scatterplot(x = X_comp, y = y_comp)

for i, (x_annot, y_annot) in enumerate(zip(X_comp, y_comp)):
    plt.annotate(data.columns[i], (x_annot, y_annot))

plt.show()
```
