```python
import pandas as pd

data = pd.read_csv("data.csv", index_col=0)
print(data[0:5])
```

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=3)
model.fit(data)
pred = model.predict(data)

data['グループ名'] = pred
print(data)
```

```python
print(data.groupby('グループ名').mean())
```
