```python
import numpy as np
import pandas as pd

data_past = pd.read_csv("./user_data.csv")
data_future = pd.read_csv("./user_data_future.csv")
```

```python
X_name = ["sex", "student", "stay time"]

X = data_past[X_name]
y = data_past["registration"]
```

```python
import statsmodels.api as sm
from sklearn.linear_model import LogisticRegression

model = sm.Logit(y, sm.add_constant(X))
result = model.fit(disp = 0)
result.summary()
```

```python
result.predict(sm.add_constant(data_future[X_name]))
```
